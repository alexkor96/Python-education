```python
import pandas as pd
```

```python
# читаем файл california_housing_test.csv
df = pd.read_csv('california_housing_test.csv')
```

```python
#кол-во строк и столбцов
df.shape
```

```python
# Тип данных столбцов
df.dtypes
```

```python
#Показать longitude где housing_median_age < 35
df.loc[df['housing_median_age'] < 35 , ['longitude']]
```

```python
#Показать столбцы longitude и latitude
df[['longitude', 'latitude']]
```

```python
#Выбрать данные где total_rooms < 2000 и population > 1500
df[(df['total_rooms'] > 2000) & (df['population'] > 1500)]
```

```python
print(df['population'].max())
```

```python
df.loc[df['median_income'] > 3, ['longitude']]['longitude'].max()
```

```python
df[(df['households']==df['households'].min())]['total_bedrooms'].max()
```
